```python
import csv
import requests
import shutil

import pandas as pd
```

```python
# Upload and read the csv files from the github repo
df = pd.read_csv("https://raw.githubusercontent.com/HelenG123/aeye-alliance/master/Labelled%20Data/ting_yi_new_dataset.csv")

print(df.shape)
df.head()
```

```python
with open("braille.csv", 'w', encoding='utf-8', newline='') as csvfile:
    writer = csv.writer(csvfile)

    writer.writerow(["Image", "Label"])

    for index, row in df.iterrows():
        url = row['Labeled Data']
        label = row['Label'][11]
        filename = url.split('/')[-1].split('?')[0]

        r = requests.get(url, stream=True)
        if r.status_code == 200:
            r.raw.decode_content = True

            with open(f"data/{filename}", 'wb') as f:
                shutil.copyfileobj(r.raw, f)
            
            writer.writerow([filename, label])
            print(f"{index:04}: Image Successfully Downloaded --> {filename}")
        else:
            with open("failed.csv", '+a', encoding='utf-8', newline='') as f:
                f_writer = csv.writer(f)
                writer.writerow([url, label])
            
            print(f"{index:04}: Could not download image --> {filename}")
```
